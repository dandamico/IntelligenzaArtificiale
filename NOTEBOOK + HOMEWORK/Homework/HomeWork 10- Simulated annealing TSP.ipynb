# Tweak Stato

```python
def tweak(sol):

    #lista = []
    t = len(sol)

    i = random.randint(0, t-1)
    j = random.randint(0, t-1)

    # for i in range(0, t-1):
    #     for j in range(i+1, t):
    #         buffer = copy.deepcopy(sol)
    #         temp = buffer[i]
    #         buffer[i] = buffer[j]
    #         buffer[j] = temp
    #         energia_successore = energia(buffer)
    #         lista.append((buffer, energia_successore))

    buffer = copy.deepcopy(sol)
    temp = buffer[i]
    buffer[i] = buffer[j]
    buffer[j] = temp
    #lista.sort(key=lambda x: x[1])  # ordiniamo i successori in base alla loro valutazione eval_stato

    #return(lista)
    return buffer
```

# Calcolo dell'energia (lunghezza percorso)

```python
def energia(stato):

    lunghezza_percorso = 0
    for i in range(len(stato)):
        l = connections[stato[i]]
        if i < (len(stato) - 1):
            nodo_succ = stato[i + 1]
        else:
            nodo_succ = stato[0]
        for (successore, distanza) in l:
            if successore == nodo_succ:
                lunghezza_percorso += distanza
                break
    return lunghezza_percorso
```

# Algoritmo Simulated Annealing - TSP

```python
def simulated_annealing(stato_iniziale):
    
    # impostazione dello stato iniziale
    current = stato_iniziale
    current_energy = energia(current)
    
    # inizializzazione best
    best = current
    best_energy = current_energy

    temperature = TEMPERATURA_INIZIALE

    while temperature > TEMPERATURA_FINALE: # si puo aggiungere and best_energy!=0 ma inutile(non arriva mai a 0)
        print("TEMPERATURA: "),
        print(temperature)
        for step in range(0,STEPS_PER_CHANGE):
            useNew = False
            #neighbours = tweak(current)
            #next_state = neighbours[0][0]
            #next_state_energy = neighbours[0][1]
            next_state = tweak(current)
            next_state_energy = energia(next_state)
            if next_state_energy < current_energy:
                useNew = True
            else:
                delta = next_state_energy - current_energy
                metropolis = math.exp(-delta/temperature)
                test = random.random()
                if (test < metropolis):
                    useNew = True
                    
            # se abbiamo deciso di accettare il nuovo stato:   
            if useNew is True:
                # impostalo come stato ed energia correnti
                current = next_state
                current_energy = next_state_energy
            
                # se è anche il migliore aggiorna best
                if current_energy < best_energy:
                    best = current
                    best_energy = current_energy

        print("best_energy="),
        print(best_energy)

        print("---------------------")
        
        # diminuisci la temperatura, senza mai arrivare a zero
        temperature = temperature * ALFA
        
    return(best + [best[0]], best_energy)
```

## Esecuzione dell'algoritmo

```python
# Impostazione parametri

TEMPERATURA_INIZIALE = 50
TEMPERATURA_FINALE = 0.2
ALFA = 0.99
STEPS_PER_CHANGE = 100
```

```python
# connessioni tra città con relative distanze

connections = {}

connections['A'] = [['B', 4], ['C', 2], ['D',  1], ['E',  2]]
connections['B'] = [['A', 4], ['C', 1], ['D', 2], ['E', 3]]
connections['C'] = [['A', 2], ['B', 1], ['D', 3], ['E',  2]]
connections['D'] = [['A', 1], ['B', 2], ['C',  3], ['E',  2]]
connections['E'] = [['A', 2], ['B',  3], ['C',  2], ['D',  2]]
```

```python
stato = ['A', 'B', 'C', 'D', 'E']
```

```python
soluzione = simulated_annealing(stato)
```
