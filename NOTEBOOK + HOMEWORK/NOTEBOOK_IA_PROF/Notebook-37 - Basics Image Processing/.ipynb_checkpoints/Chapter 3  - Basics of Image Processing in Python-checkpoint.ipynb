### 3.2.3. Converting an Image to Grayscale

```python
import cv2

panda_img_temp = cv2.imread("panda_image.jpg")

panda_img = cv2.cvtColor(panda_img_temp, cv2.COLOR_BGR2GRAY)
plt.imshow(panda_img);
```

### 3.2.4. Cropping an Image

```python
import cv2

panda_img = cv2.imread("panda_image.jpg")
y=200
x=200
h=100
w=200

crop_img = panda_img[y:y+h, x:x+w]
plt.imshow(crop_img);
```

### 3.2.5.  Resizing an Image with 

```python
import cv2

panda_img_temp = cv2.imread("panda_image.jpg")

resized_image = cv2.resize(panda_img_temp, (100, 50)) 

plt.imshow(resized_image);
```

```python
import cv2

panda_img_temp = cv2.imread("panda_image.jpg")

resized_image = cv2.resize(panda_img_temp, (0,0), fx=0.5, fy=0.5) 

plt.imshow(resized_image)

height, width, channels = resized_image.shape 

print(height)
print(width)
print(channels)
```

### 3.2.6. Applying Filters


```python
import cv2

panda_img_temp = cv2.imread("panda_image.jpg")

panda_img = cv2.blur(panda_img_temp, (10,10),0)

plt.imshow(panda_img);
```

```python
import cv2

panda_img_temp = cv2.imread("panda_image.jpg")

panda_img = cv2.Canny(panda_img_temp, 80, 200)

plt.imshow(panda_img);
```

### 3.2.7. Drawing Shapes

#### Drawing a Line

```python
pandas_img = cv2.imread("panda_image.jpg")

cv2.line(
              pandas_img,
              pt1 = (100,100), pt2 = (100,200),
              color = (255,0,0),
              thickness = 10
            )
plt.imshow(pandas_img);
```

#### Drawing a Rectangle

```python
pandas_img = cv2.imread("panda_image.jpg")

cv2.rectangle(
              pandas_img,
              pt1 = (100,100), pt2 = (200,200),
              color = (19,52,100),
              thickness = 5
            )
plt.imshow(pandas_img);
```

#### Drawing a Circle

```python
pandas_img = cv2.imread("panda_image.jpg")

cv2.circle(
              pandas_img,
              center = (200,200),
              radius = 100,
              color = (255,5,210),
              thickness = 10
            )
plt.imshow(pandas_img);
```

### 3.3.2. Saving an Image in a Different Format

### 3.3.3. Changing Image Color

```python
from PIL import Image
panda_img = Image.open(r'panda_image.jpg')

panda_img_gs = panda_img.convert('L')
display(panda_img_gs)
```

### 3.3.4. Changing Image Orientation

```python
from PIL import Image
panda_img = Image.open(r'panda_image.jpg')

panda_img_ud = panda_img.transpose(Image.ROTATE_90)
display(panda_img_ud )
```

```python
from PIL import Image
panda_img = Image.open(r'panda_image.jpg')

panda_img_ud  = panda_img.transpose(Image.FLIP_LEFT_RIGHT)
display(panda_img_ud)
```

```python
from PIL import Image
panda_img = Image.open(r'panda_image.jpg')

panda_img_ud  = panda_img .transpose(Image.FLIP_TOP_BOTTOM)
display(panda_img_ud)
```

### 3.3.5. Cropping an Image

```python
from PIL import Image
panda_img = Image.open(r'panda_image.jpg')

crop_dimensions = (100,100,450,450)
panda_img_ud = panda_img.crop(crop_dimensions)
display(panda_img_ud)
```

### 3.3.6. Modifying an Image Area

```python
from PIL import Image
panda_img = Image.open(r'panda_image.jpg')

crop_dimensions = (100,100,450,450)
panda_cropped = panda_img.crop(crop_dimensions)
panda_cropped = panda_cropped.convert('L')

panda_img.paste(panda_cropped, crop_dimensions)

display(panda_img)
```

### 3.3.7. Resizing an Image

```python
from PIL import Image
panda_img = Image.open(r'panda_image.jpg')

new_size = (300,200)
panda_img_ud  = panda_img.resize(new_size) 
display(panda_img_ud)
```

### 3.3.8. Applying Filters to Images

```python
from PIL import Image, ImageFilter
from PIL import Image
panda_img = Image.open(r'panda_image.jpg')


panda_img_ud  = panda_img.filter(ImageFilter.EDGE_ENHANCE_MORE) 
display(panda_img_ud)
```

```python
from PIL import Image, ImageFilter
from PIL import Image
panda_img = Image.open(r'panda_image.jpg')


panda_img_ud  = panda_img.filter(ImageFilter.BLUR) 
display(panda_img_ud)
```

## Exercise 3.2

Import the "lion_image.jpg" to your Python script using the Pillow library and perform the following tasks:
    
1. Convert the image to Grayscale
2. Crop the 500 X 500 pixel image from 100 pixels from top and 100 pixels from left.
3. Invert the image
4. Apply edge enhance filter
5. Display the Image.


**Solution:**

```python
from PIL import Image, ImageFilter
from PIL import Image
lion_img = Image.open(r'E:/Datasets/lion_image.jpg')

lion_img_gs = lion_img.convert('L')

crop_dimensions = (100,100,500,500)
lion_img_ud = lion_img_gs.crop(crop_dimensions)

lion_img_ud  = lion_img_ud .transpose(Image.FLIP_TOP_BOTTOM)

lion_img_ud  = lion_img_ud.filter(ImageFilter.EDGE_ENHANCE_MORE) 

display(lion_img_ud)
```
